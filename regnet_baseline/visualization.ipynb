```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_gpu_1 = pd.read_csv("training_stats_GPU_1_batchsize_35.csv")
data_gpu_2 = pd.read_csv("training_stats_GPU_2_batchsize_35.csv")
data_gpu_4 = pd.read_csv("training_stats_GPU_4_batchsize_35.csv")
data_gpu_8 = pd.read_csv("training_stats_GPU_8_batchsize_35.csv")
data_gpu_16 = pd.read_csv("training_stats_GPU_16_batchsize_35.csv")
```

```python
mean_1 = data_gpu_1['time (ms)'].mean()
std_1 = data_gpu_1['time (ms)'].std()
mean_2 = data_gpu_2['time (ms)'].mean()
std_2 = data_gpu_2['time (ms)'].std()
mean_4 = data_gpu_4['time (ms)'].mean()
std_4 = data_gpu_4['time (ms)'].std()
mean_8 = data_gpu_8['time (ms)'].mean()
std_8 = data_gpu_8['time (ms)'].std()
mean_16 = data_gpu_16['time (ms)'].mean()
std_16 = data_gpu_16['time (ms)'].std()
```

```python
plt.title("1 GPU, Batchsize: 77, All epochs")
plt.xlabel("Time per image (ms)")
plt.ylabel("Number of images")
plt.hist(data_gpu_1['time (ms)'],range=(mean_1-0.5*std_1,mean_1+0.5*std_1),bins=30)
plt.show()
```

```python
plt.title("1 GPU, Batchsize: 77, First epoch")
plt.xlabel("Time per image (ms)")
plt.ylabel("Number of images")
plt.hist(data_gpu_1['time (ms)'][data_gpu_1['epoch']==0.0],range=(mean_1-0.5*std_1,mean_1+0.5*std_1),bins=30)
plt.show()
```

```python
plt.title("1 GPU, Batchsize: 77, Second epoch")
plt.xlabel("Time per image (ms)")
plt.ylabel("Number of images")
plt.hist(data_gpu_1['time (ms)'][data_gpu_1['epoch']==1.0],range=(mean_1-0.5*std_1,mean_1+0.5*std_1),bins=30)
plt.show()
```
